# Importando SQL
SQL - Structured Query Languade - Linguagem de consulta estruturada

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:') #indica que estou usando banco de dados sqlite, memory- usando banco de dados na memoria local
```

```python
dados_cliente_banco = pd.read_csv('/content/clientes_banco.csv')
```

```python
dados_cliente_banco.head()
```

```python
dados_cliente_banco.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine) # para ler uma consulta SQL
```

```python
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False) #con indica a conexão ao engine criado anteriormente
```

```python
pd.read_sql_table('empregados', engine) #para ler uma tabela completa
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual']) #para ler uma tabela completa
```

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
# após a leitura dos registros da tabela clientes, é fundamental iniciar aimportando a função text do SQLAlchemy
# função essencial para criar e executar expressões SQL literais dentro do código python
from sqlalchemy import text

query = text('DELETE FROM clientes WHERE ID_Cliente=5008804')
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()  # importante para efetivar a alteração
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = text('UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808')
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()  # importante para efetivar a alteração
```

```python
pd.read_sql_table('clientes', engine)
```
